#  Python Project :- Calculator

```python
from tkinter import *
#Blank string to store value from user
expression=""

#def for color
def colour():
    from tkinter import colorchooser
    tup=colorchooser.askcolor()
    frm.configure(bg=tup[1])

#for input no. in entry 1
def press(num):
    global expression
    expression=expression+str(num)
    equation1.set(expression)

    
#for finding the square of a no.
def numsquare():
    global expression
    n=int(expression)
    equation2.set(n*n)

def change_color1(event):
    b20.configure(bg="green")

def change_color2(event):
    b20.configure(bg="#ABB7B7")
    
    
#for clearing both entry1 and entry2
def clear():
    global expression
    expression=""
    equation1.set("")
    equation2.set("")
#To evaluate the solution using try and except block
def but20(): 
    try: 
        global expression 
        total = str(eval(expression)) 
 
        equation2.set(total) 
        expression = "" 
 
    except: 
        equation2.set(" error ") 
        expression = "" 
    
win=Tk()
win.title("Calculator")
win.configure(bg="#ABB7B7")
win.geometry("372x577")
win.resizable(width=False,height=False)
equation1=StringVar()
equation2=StringVar()

#for input data
e1=Entry(win,textvariable=equation1,font=('arial',20,'bold'),bd=6,width=23)
e1.place(x=5,y=15)
#for output data
e2=Entry(win,textvariable=equation2,font=('arial',20,'bold'),bd=6,width=23)
e2.place(x=5,y=67)

#Frame for buttons
frm=Frame(win)
frm.configure(bg="#799351")
frm.place(x=2,y=150,width=368,height=425)

#For buttons 

b1=Button(frm,text="Colour",font=('arial',10,'bold'),bd=5,width=9,height=4,command=colour)
b1.grid(row=0,column=0)


b2=Button(frm,text="(",font=('arial',10,'bold'),bd=5,width=9,height=4,command=lambda: press('('))
b2.grid(row=0,column=1)

b3=Button(frm,text=")",font=('arial',10,'bold'),bd=5,width=9,height=4,command=lambda: press(')'))
b3.grid(row=0,column=2)

b4=Button(frm,text="+",font=('arial',10,'bold'),bd=5,width=9,height=4,command=lambda: press('+'))
b4.grid(row=0,column=3)

b5=Button(frm,text="1",font=('arial',10,'bold'),bd=5,width=9,height=4,command=lambda: press(1))
b5.grid(row=1,column=0)

b6=Button(frm,text="2",font=('arial',10,'bold'),bd=5,width=9,height=4,command=lambda: press(2))
b6.grid(row=1,column=1)

b7=Button(frm,text="3",font=('arial',10,'bold'),bd=5,width=9,height=4,command=lambda: press(3))
b7.grid(row=1,column=2)

b8=Button(frm,text="-",font=('arial',10,'bold'),bd=5,width=9,height=4,command=lambda: press('-'))
b8.grid(row=1,column=3)

b9=Button(frm,text="4",font=('arial',10,'bold'),bd=5,width=9,height=4,command=lambda: press(4))
b9.grid(row=2,column=0)

b10=Button(frm,text="5",font=('arial',10,'bold'),bd=5,width=9,height=4,command=lambda: press(5))
b10.grid(row=2,column=1)

b11=Button(frm,text="6",font=('arial',10,'bold'),bd=5,width=9,height=4,command=lambda: press(6))
b11.grid(row=2,column=2)

b12=Button(frm,text="*",font=('arial',10,'bold'),bd=5,width=9,height=4,command=lambda: press('*'))
b12.grid(row=2,column=3)

b13=Button(frm,text="7",font=('arial',10,'bold'),bd=5,width=9,height=4,command=lambda: press(7))
b13.grid(row=3,column=0)

b14=Button(frm,text="8",font=('arial',10,'bold'),bd=5,width=9,height=4,command=lambda: press(8))
b14.grid(row=3,column=1)

b15=Button(frm,text="9",font=('arial',10,'bold'),bd=5,width=9,height=4,command=lambda: press(9))
b15.grid(row=3,column=2)

b16=Button(frm,text="/",font=('arial',10,'bold'),bd=5,width=9,height=4,command=lambda: press('/'))
b16.grid(row=3,column=3)

b17=Button(frm,text="X*2",font=('arial',10,'bold'),bd=5,width=9,height=4,command=numsquare)
b17.grid(row=4,column=0)

b18=Button(frm,text="0",font=('arial',10,'bold'),bd=5,width=9,height=5,command=lambda: press(0))
b18.grid(row=4,column=1)

b19=Button(frm,text="Clear",font=('arial',10,'bold'),bd=5,width=9,height=4,command=clear)
b19.grid(row=4,column=2)

b20=Button(frm,text="=",font=('arial',12,'bold'),bd=5,width=9,height=4,command=but20)
b20.grid(row=4,column=3)
b20.bind('<Enter>',change_color1)
b20.bind('<Leave>',change_color2)

e1.focus_set()
win.mainloop()
```
